```python
import pandas as pd

# Load the dataset
file_path = '/Users/nisha/Desktop/heart_attack_prediction_dataset.csv'
df = pd.read_csv(file_path)

# Display basic information about the dataset
print("Original Dataset Info:")
print(df.info())

# Display first few rows of the dataset
print("\nOriginal Dataset:")
print(df.head())

# Handling Missing Values
# For numerical columns, you can impute missing values with mean or median
numeric_cols = df.select_dtypes(include='number').columns.tolist()
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)  # Filling missing values with median

# For categorical columns, you can impute missing values with mode
categorical_cols = df.select_dtypes(include='object').columns.tolist()
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)  # Filling missing values with mode

# Handling Outliers
# Implement outlier handling techniques such as Z-score or IQR method
# Define functions or methods to handle outliers in specific columns

# For example, using Z-score method to handle outliers in a specific numerical column 'Feature_X'
def handle_outliers_zscore(column):
    z_scores = (column - column.mean()) / column.std()
    abs_z_scores = abs(z_scores)
    outliers = abs_z_scores > 3  # Define your threshold for outliers
    column[outliers] = column.median()  # Replace outliers with median value

handle_outliers_zscore(df['Feature_X'])

# Save the cleaned dataset to a new file or overwrite the existing one
cleaned_file_path = 'cleaned_physical_dataset.csv'
df.to_csv(cleaned_file_path, index=False)

# Display basic information about the cleaned dataset
print("\nCleaned Dataset Info:")
print(df.info())

# Display first few rows of the cleaned dataset
print("\nCleaned Dataset:")
print()
```
