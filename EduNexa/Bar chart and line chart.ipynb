```python
import matplotlib.pyplot as plt

quarters = ['Q1', 'Q2', 'Q3', 'Q4']
sales = [350, 420, 390, 480]

# Bar chart for sales across quarters
plt.figure(figsize=(8, 5))
plt.bar(quarters, sales, color='skyblue')
plt.xlabel('Quarters')
plt.ylabel('Sales')
plt.title('Sales Performance Across Quarters')
plt.show()

# Line chart for sales trend
plt.figure(figsize=(8, 5))
plt.plot(quarters, sales, marker='o', linestyle='-', color='orange')
plt.xlabel('Quarters')
plt.ylabel('Sales')
plt.title('Sales Trend Across Quarters')
plt.grid(True)  # Adding gridlines for better visualization
plt.show()
```
